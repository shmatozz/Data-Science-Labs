```python
data = data[(data["y"] < 20) & (data["z"] < 10)]
```

```python
data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
data.hist(bins=50, figsize=(8,8));
```

```python
sns.boxplot(y=data['price'], x=data['cut'], hue=data['color'])
```

```python
data_cheap = data[data['price'] < 5000]
data_expencive = data[data['price'] >= 5000]
```

```python
sns.boxplot(y=data_cheap['price'], x=data_cheap['cut'], hue=data_cheap['color'])
```

```python
sns.boxplot(y=data_expencive['price'], x=data_expencive['cut'], hue=data_expencive['color'])
```

```python
mpl.scatter(data.carat, data.price)
```

```python
def find_iqr(x):
  return np.percentile(x, [75, 25])

data[['carat', 'depth', 'price']].apply(find_iqr)
```

```python
corr = data.corr(numeric_only=True)
corr
```
